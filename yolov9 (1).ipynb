```python
import gradio as gr
import os
import subprocess
from PIL import Image as PILImage
import glob

# Define paths
HOME = os.getcwd()
weights_path = "/content/yolov9/runs/train/exp/weights/best.pt"  # Ensure this path is correct
output_dir = f"{HOME}/yolov9/runs/detect"

# Function to run YOLOv9 detection and capture output as text
def detect_objects(image):
    # Save the uploaded image locally
    image_path = f"{HOME}/uploaded_image.jpg"
    image.save(image_path)

    # Command to run YOLOv9 detection on the uploaded image
    command = f"python detect.py --weights {weights_path} --conf 0.1 --source {image_path} --device 0"

    # Run the command and capture the output
    process = subprocess.Popen(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    stdout, stderr = process.communicate()

    # Decode the output
    stdout_str = stdout.decode('utf-8')
    stderr_str = stderr.decode('utf-8')

    # Check for errors
    if stderr_str:
        return f"Error during detection: {stderr_str}"

    # Extract relevant parts from the output
    result_lines = []

    # Split the stdout into lines
    for line in stdout_str.split('\n'):
        if "image" in line and "crack" in line:
            result_lines.append(line.strip())

    # Add the path to the saved results
    result_lines.append(f"Results saved to {output_dir}/exp9")

    # Check if any results files are available
    result_files = glob.glob(f"{output_dir}/exp9/*.txt")
    if result_files:
        for result_file in result_files:
            with open(result_file, 'r') as file:
                result_lines.append(file.read())
    else:
        result_lines.append("No result files found.")

    # Return the detection results as text
    return "\n".join(result_lines) if result_lines else "No detection results. Please check your setup or try another image."

# Create the Gradio interface
interface = gr.Interface(
    fn=detect_objects,
    inputs=gr.Image(type="pil", label="Upload an Image"),
    outputs="text",
    title="YOLOv9 Object Detection - Text Output",
    description="Upload an image and YOLOv9 will detect objects. The result will show detected objects and confidence levels."
)

# Launch the interface
interface.launch()
```

```python
import gradio as gr
import subprocess
import os
from PIL import Image

# Define paths
weights_path = "/content/yolov9/runs/train/exp/weights/best.pt"  # Update with your weights path
results_base_dir = "/content/yolov9/runs/detect"
uploaded_image_path = "/content/yolov9/uploaded_image.jpg"

def detect_objects(image):
    # Save the uploaded image locally
    image.save(uploaded_image_path)

    # Run YOLOv9 detection
    command = f"python /content/yolov9/detect.py --weights {weights_path} --conf 0.1 --source {uploaded_image_path} --device 0"
    process = subprocess.Popen(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    stdout, stderr = process.communicate()

    # Print stdout and stderr for debugging
    print(f"Detection stdout: {stdout.decode('utf-8')}")
    if stderr:
        print(f"Detection stderr: {stderr.decode('utf-8')}")

    # Get the latest experiment directory
    def get_latest_exp_dir(base_dir):
        exp_dirs = [os.path.join(base_dir, d) for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))]
        if not exp_dirs:
            return None
        return max(exp_dirs, key=os.path.getmtime)

    latest_exp_dir = get_latest_exp_dir(results_base_dir)
    if not latest_exp_dir:
        print("No experiment directories found.")
        return Image.new('RGB', (640, 640), color='white')  # Return a blank image if no exp dir

    output_image_path = os.path.join(latest_exp_dir, "uploaded_image.jpg")

    # Print the path to debug
    print(f"Checking for result image at: {output_image_path}")

    # Check if the result image is saved
    if os.path.exists(output_image_path):
        result_image = Image.open(output_image_path)
        return result_image
    else:
        print("Result image not found.")
        return Image.new('RGB', (640, 640), color='white')  # Return a blank image if no result image

# Gradio interface
interface = gr.Interface(
    fn=detect_objects,
    inputs=gr.Image(type="pil", label="Upload an Image"),
    outputs="image",
    title="YOLOv9 Object Detection",
    description="Upload an image and get object detection results."
)

interface.launch()
```
